Importing Libaries

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("/home/syntaxerror/Desktop/Machine Learning/Score-Prediction/Student - Score.csv")
df.head()
```

```python
x = df['Hours'].values
y = df['Scores'].values

x1 = x.reshape(-1,1)
```

Fitting The Model

```python
model = LinearRegression()
model.fit(x1,y)
```

Plotting Function

```python
def regression_plot(x,y,model,title):
    
    plt.figure(figsize=(14,7))
    plt.title(title)

    x_line = np.array([np.min(x), np.max(x)]).reshape(-1,1)
    y_line = model.predict(x_line)

    plt.scatter(x,y,label="Data Points",color='r')
    plt.plot(x_line,y_line,label="Regression Line",color='b')

    plt.legend()
    plt.show()
```

```python
regression_plot(x1,y,model,"Linear Regression")
```

Error Calculation

```python
y_pred = model.predict(x1)
print(f"Mean Squared Error : {mean_squared_error(y,y_pred):.2f}")
```

Predicting Score Where Time = 9.25hr

```python
new_value = np.array([9.25])
new_value = new_value.reshape(-1,1)
prediction = model.predict(new_value)
print(f"Prediction : {prediction[0]:.2f}")
```
